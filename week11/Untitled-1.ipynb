```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
import seaborn as sns

# 1. Load the data
df = pd.read_excel("C:/Users/Nguyen Huyen Thuong/Downloads/Data_MidTerm.xlsx")

# 2. Inspect key columns
print("Sample of Recency, Frequency, and Monetary clusters:")
print(df[['RecencyCluster', 'FrequencyCluster', 'MonetaryCluster']].head(), '\n')

# 3. Visualize distribution of existing clusters
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.countplot(x='RecencyCluster', data=df, ax=axes[0])
axes[0].set_title('Recency Cluster Counts')
sns.countplot(x='FrequencyCluster', data=df, ax=axes[1])
axes[1].set_title('Frequency Cluster Counts')
sns.countplot(x='MonetaryCluster', data=df, ax=axes[2])
axes[2].set_title('Monetary Cluster Counts')
plt.tight_layout()
plt.show()

# 4. Scatter plot of clusters
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='RecencyCluster', 
    y='FrequencyCluster', 
    hue='MonetaryCluster', 
    data=df, 
    palette='tab10',
    alpha=0.6
)
plt.title('Recency vs Frequency by Monetary Cluster')
plt.show()

# 5. Prepare features and target
X = df[['RecencyCluster', 'FrequencyCluster']]
# Shift MonetaryCluster to zero-based labels for multi-class
y = df['MonetaryCluster'].astype(int) - 1  
n_classes = y.nunique()

# 6. Train/Test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# 7. Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 8. Define models
models = {
    'Logistic Regression': LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000),
    'XGBoost': XGBClassifier(use_label_encoder=False, objective='multi:softmax', num_class=n_classes, eval_metric='mlogloss'),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(50,), max_iter=1000)
}

# 9. Train, predict, and evaluate
results = {}
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    preds = model.predict(X_test_scaled)
    
    print(f"\n=== {name} Classification Report ===")
    print(classification_report(y_test, preds, digits=4))
    
    cm = confusion_matrix(y_test, preds)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{name} Confusion Matrix')
    plt.xlabel('Predicted Class')
    plt.ylabel('True Class')
    plt.show()
    
    results[name] = classification_report(y_test, preds, output_dict=True)

# 10. Summary of performance metrics
summary = pd.DataFrame([
    {
        'Model': name,
        'Accuracy': results[name]['accuracy'],
        **{f'Precision_Class{cls}': results[name][str(cls)]['precision'] for cls in range(n_classes)},
        **{f'Recall_Class{cls}': results[name][str(cls)]['recall'] for cls in range(n_classes)},
        **{f'F1_Class{cls}': results[name][str(cls)]['f1-score'] for cls in range(n_classes)}
    }
    for name in results
])

print("Overall Performance Summary:")
display(summary.set_index('Model'))
```

```python
import pandas as pd
from itertools import combinations

# Định nghĩa transactions
transactions = [
    ['Student', 'Teach', 'School'],
    ['Student', 'School'],
    ['Teach', 'School', 'City', 'Game'],
    ['Baseball', 'Basketball'],
    ['Basketball', 'Player', 'Spectator'],
    ['Baseball', 'Coach', 'Game', 'Team'],
    ['Basketball', 'Team', 'City', 'Game']
]

# Thiết lập ngưỡng
min_support = 0.3  # 30%
min_confidence = 0.3  # 30%
n_transactions = len(transactions)
min_support_count = min_support * n_transactions

# 1. Tính support count cho từng item
item_counts = {}
for txn in transactions:
    for item in txn:
        item_counts[item] = item_counts.get(item, 0) + 1

# 2. Lọc frequent 1-itemsets
freq1 = {item for item, count in item_counts.items() if count >= min_support_count}

# 3. Tìm frequent 2-itemsets từ freq1
freq2_counts = {}
for combo in combinations(freq1, 2):
    count = sum(1 for txn in transactions if all(i in txn for i in combo))
    if count >= min_support_count:
        freq2_counts[combo] = count

# 4. Sinh luật kết hợp từ các frequent 2-itemsets
rules = []
for itemset, support_count in freq2_counts.items():
    support = support_count / n_transactions
    for i in range(1, len(itemset)):
        for antecedent in combinations(itemset, i):
            consequent = tuple(set(itemset) - set(antecedent))
            antecedent_count = sum(1 for txn in transactions if all(a in txn for a in antecedent))
            confidence = support_count / antecedent_count
            if confidence >= min_confidence:
                rules.append({
                    'antecedent': antecedent,
                    'consequent': consequent,
                    'support': round(support, 3),
                    'confidence': round(confidence, 3)
                })

# 5. Xuất kết quả
print("Frequent 1-itemsets (support >= 30%):")
print([{item: item_counts[item]} for item in freq1])

print("\nFrequent 2-itemsets (support >= 30%):")
print([{combo: count} for combo, count in freq2_counts.items()] or "None")

print("\nAssociation Rules (confidence >= 30%):")
print(rules or "None")

```
